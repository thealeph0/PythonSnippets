Write a snippet of code (not a function, just the lines of code necessary) to prompt the user for a driver speed and then prints out the classification based on that wind speed.

Use the int and input functions to prompt the user for the driving speed. You may assume the user correctly enters an int. Any value less than or equal to 0 results in the error message shown below. 

Validation: Re-prompt the user if they enter a negative wind speed.

The code snippet prints the classification of the given wind speed based on the following criteria. Note the [ symbol means inclusive and the ) symbol means exclusive.

<ol>
[0, 25) -> Calm <p>
[25, 35) -> Standard <p>
[35, 50) -> Fast <p>
[50, 65) -> Very Quickly <p>
>= 65 -> Violent <p>
</ol>

Restrictions: The only Python functions you may call are the input, int, and print functions. Avoid doing unnecessary calculations.

```python
speed = int(input("Enter your speed: "))

#input validation
while speed <=0:
    print("Invalid speed, try again")
    speed = int(input("Enter your speed: "))

if speed < 25:
    state = 'calm'
elif speed < 35:
    state = 'standard'
elif speed < 50:
    state = 'fast'
elif speed < 65:
    state = 'very quickly'
else:
    state = 'violent'
    
print("You are driving: ", state)
```
